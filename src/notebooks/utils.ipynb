# CSV

## Changing columns value in csv

```python
def change_columns_csv(path):
    import csv
    r = csv.reader(open(path)) # Here your csv file
    lines = list(r)
    lines[0]
    import copy
    for line in lines:
        img_name = copy.deepcopy(line[0]) 
        line[0] = f"iNat-1000/{img_name}"
        print(line[0])
    writer = csv.writer(open('../data/inputs/iNat1000/iNat-1000_goodpath.csv', 'w'))
    writer.writerows(lines)

change_columns_csv('../data/inputs/iNat1000/iNat-1000.csv')
```
