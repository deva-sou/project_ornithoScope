```python
def del_random_element(df,caped_number):
    size = len(df)
    np.random.seed(10)
    remove_n = size-caped_number
    print(remove_n)
    drop_indices = np.random.choice(df.index, remove_n, replace=False)
    print(drop_indices)
    df_subset = df.drop(drop_indices)
    return df_subset
```

```python
def cap_label_number(_df, caped_number):
    df = _df.copy()
    for specie in list_of_species:
        if df[df['label'] == specie].shape[0] >= caped_number:
            print('too big')
            df[df['label'] == specie] = del_random_element(df[df['label'] == specie],caped_number)
            print('done')
        else:
            print('ok')
    return df
```
