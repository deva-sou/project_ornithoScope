```python
df_iNat = pd.read_csv('../../data/inputs/iNat1000/iNat1000_goodpath.csv',
                       names = ["path","x1","x2","y1","y2","label","height","width"])
df_iNat
```

```python
df_cap1000 = pd.read_csv('../../data/inputs/input_caped/input_train_caped_1000.csv',
                       names = ["path","x1","x2","y1","y2","label","height","width"])
df_cap1000
```

```python
list_labels_iNat = list(df_iNat["label"].unique())
list_labels_cap300 = list(df_cap1000["label"].unique())
print(list(set(list_labels_iNat) - set(list_labels_cap300)))
```

```python
def print_number_of_object_pr_classes(list_labels,df):
    for label in list_labels_cap300:
        print(f"{label} {len(df[df['label'] == label])}")
print_number_of_object_pr_classes(list_labels_cap300,df_cap1000)
```

```python
def fill_df_caped(df_caped,df_iNat,filled_to):
    list_df_to_add = [df_caped]
    for label in list_labels_cap300:
        if len(df_caped[df_caped['label'] == label]) < filled_to: 
            number_to_add = filled_to - len(df_caped[df_caped['label'] == label])
            df_to_add = df_iNat[df_iNat['label'] == label].iloc[:number_to_add]
            list_df_to_add.append(df_to_add)
    df = pd.concat(list_df_to_add, ignore_index = True, axis = 0)
    return df
        
_ = fill_df_caped(df_cap1000, df_iNat,1000)
_
```

```python
print_number_of_object_pr_classes(list_labels_cap300,_)
```

```python
_.to_csv('../../data/inputs/input_caped/input_train_cap1000_filled_iNat.csv', header=False,index=False)
```
